### Precipitation Type Breakdown
Let's see the distribution of precipitation types in the dataset.

```python
precip_counts = df['Precip Type'].value_counts(dropna=False)
ax = precip_counts.plot(
    kind='bar', 
    color='cornflowerblue', 
    edgecolor='grey', 
    linewidth=2
)
plt.title('Precipitation Type Distribution', fontsize=14, fontweight='bold')
plt.xlabel('Precipitation Type', fontsize=12, fontweight='bold')
plt.ylabel('Count', fontsize=12, fontweight='bold')
# Annotate each bar with its value in bold, fontsize 11, and black color, with a clear gap above the bar
for i, v in enumerate(precip_counts):
    ax.text(i, v + 0.03 * max(precip_counts), str(v), ha='center', va='bottom', fontsize=11, fontweight='bold', color='black')
plt.ylim(0, max(precip_counts) * 1.10)  # Add extra space above the tallest bar
plt.show()
precip_counts
```

### Time Series Trend: Average Temperature Over Time
Let's analyze how the average temperature changes over time.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('weatherHistory/weatherHistory.csv')

# Print the column names
print("Original Columns:", df.columns)

# Strip any leading/trailing spaces in column names
df.columns = df.columns.str.strip()
print("Stripped Columns:", df.columns)

# Ensure 'Formatted Date' column exists
if 'Formatted Date' in df.columns:
    # Convert to datetime with timezone support
    df['Formatted Date'] = pd.to_datetime(df['Formatted Date'], utc=True, errors='coerce')
    # Set as index
    df.set_index('Formatted Date', inplace=True)
    # Confirm index is datetime
    if isinstance(df.index, pd.DatetimeIndex):
        # Resample by month-end and plot
        monthly_avg = df.resample('M')['Temperature (C)'].mean()
        fig, ax = plt.subplots(figsize=(12, 6))
        # Plot the line and points with bold effect, but use smaller marker size
        ax.plot(monthly_avg.index, monthly_avg.values, color='tomato', linewidth=3, marker='o', markersize=4, markeredgewidth=1, markeredgecolor='black')
        # Make x and y axis labels and ticks bold
        ax.set_title('Monthly Average Temperature Over Time', fontsize=16, fontweight='bold')
        ax.set_xlabel('Date', fontsize=14, fontweight='bold')
        ax.set_ylabel('Average Temperature (Â°C)', fontsize=14, fontweight='bold')
        ax.tick_params(axis='both', labelsize=12, labelcolor='black', width=2, length=6)
        for label in ax.get_xticklabels() + ax.get_yticklabels():
            label.set_fontweight('bold')
        # Make the axis lines bold
        ax.spines['bottom'].set_linewidth(2)
        ax.spines['left'].set_linewidth(2)
        ax.spines['top'].set_linewidth(2)
        ax.spines['right'].set_linewidth(2)
        # Annotate each point with a smaller dot (no number)
        for x, y in zip(monthly_avg.index, monthly_avg.values):
            ax.plot(x, y, 'o', color='black', markersize=4)
        plt.grid(True, linestyle='--', alpha=0.6)
        plt.tight_layout()
        plt.show()
    else:
        print('Error: Index is not a DatetimeIndex. Current index type:', type(df.index))
else:
    print('Error: "Formatted Date" column not found after stripping. Current columns:', df.columns.tolist())
```

### Outlier Detection: Temperature
Let's visualize outliers in the temperature data using a boxplot.

```python
plt.figure(figsize=(22,10))
sns.boxplot(x=df['Temperature (C)'], color='gold', linewidth=2.5, notch=True, fliersize=8, boxprops=dict(alpha=0.7))

# Calculate statistics
mean_temp = df['Temperature (C)'].mean()
median_temp = df['Temperature (C)'].median()
std_temp = df['Temperature (C)'].std()

# Draw mean, median, and std lines over the boxplot
plt.axvline(mean_temp, color='blue', linestyle='--', linewidth=4, label=f'Mean: {mean_temp:.2f}')
plt.axvline(median_temp, color='green', linestyle='-', linewidth=4, label=f'Median: {median_temp:.2f}')
plt.axvline(mean_temp + std_temp, color='red', linestyle=':', linewidth=4, label=f'+1 Std: {(mean_temp + std_temp):.2f}')
plt.axvline(mean_temp - std_temp, color='red', linestyle=':', linewidth=4, label=f'-1 Std: {(mean_temp - std_temp):.2f}')

# Adjust annotation positions to keep them inside the box area
ylim = plt.ylim()
box_y = ylim[0] + (ylim[1] - ylim[0]) * 0.15  # 15% above the bottom for all annotations
plt.text(median_temp, box_y, f'Median\n{median_temp:.2f}', color='green', fontsize=22, fontweight='bold', ha='center', va='bottom', backgroundcolor='white')
plt.text(mean_temp, box_y + (ylim[1] - ylim[0]) * 0.10, f'Mean\n{mean_temp:.2f}', color='blue', fontsize=22, fontweight='bold', ha='center', va='bottom', backgroundcolor='white')
plt.text(mean_temp + std_temp, box_y, f'+1 Std\n{(mean_temp + std_temp):.2f}', color='red', fontsize=22, fontweight='bold', ha='center', va='bottom', backgroundcolor='white')
plt.text(mean_temp - std_temp, box_y, f'-1 Std\n{(mean_temp - std_temp):.2f}', color='red', fontsize=22, fontweight='bold', ha='center', va='bottom', backgroundcolor='white')

plt.title('Boxplot of Temperature (C) with Mean, Median, Â±1 Std', fontsize=28, fontweight='bold', pad=60)
plt.xlabel('Temperature (C)', fontsize=24, fontweight='bold', labelpad=20)
plt.ylabel('Count', fontsize=24, fontweight='bold', labelpad=20)
plt.xticks(fontsize=20, fontweight='bold')
plt.yticks([], [])
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.legend(fontsize=20, loc='upper right', frameon=True)
plt.tight_layout()
plt.show()
```
