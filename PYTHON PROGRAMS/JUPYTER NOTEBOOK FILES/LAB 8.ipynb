```python
import numpy as np
from scipy import stats

data = [183, 152, 178, 157, 194, 163, 144, 114, 178, 152, 118, 158, 172, 138]
mu_0 = 165

# Calculate sample mean and standard deviation
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
n = len(data)

# Calculate t-value
t_value = (sample_mean - mu_0) / (sample_std / np.sqrt(n))

# Degrees of freedom
df = n - 1

# Calculate p-value
p_value = stats.t.cdf(t_value, df)

print(f"t-value: {t_value}")
print(f"p-value: {p_value}")

# Decision
alpha = 0.05
if p_value < alpha:
    print("Reject H0: The population mean is less than 165.")
else:
    print("Fail to reject H0: There is no significant mean difference in marks.")
```

```python
import numpy as np
from scipy import stats

X1 = [13.4, 10.9, 11.2, 11.8, 14, 15.3, 14.2, 12.6, 17, 16.2, 16.5, 15.7]
X2 = [12, 11.7, 10.7, 11.2, 14.8, 14.4, 13.9, 13.7, 16.9, 16, 15.6, 16]

# Perform two-sample t-test
t_value, p_value = stats.ttest_ind(X1, X2)

print(f"t-value: {t_value}")
print(f"p-value: {p_value}")

# Decision
alpha = 0.05
if p_value < alpha:
    print("Reject H0: The effect of pollution and soil on rice paddy is not equal.")
else:
    print("Fail to reject H0: The effect of pollution and soil on rice paddy are equal.")
```

```python
import numpy as np
from scipy import stats

test1 = [23, 20, 19, 21, 18, 20, 18, 17, 23, 16, 19]
test2 = [24, 19, 22, 18, 20, 22, 20, 20, 23, 20, 18]

# Perform paired t-test
t_value, p_value = stats.ttest_rel(test1, test2)

print(f"t-value: {t_value}")
print(f"p-value: {p_value}")

# Decision
alpha = 0.05
if p_value < alpha:
    print("Reject H0: The sportsmen have benefited from the training.")
else:
    print("Fail to reject H0: The sportsmen have not benefited from the training.")
```
