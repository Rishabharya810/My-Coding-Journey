```python
s = "welcome to bennett university"

def capitalize_first_last(text):
  words = text.split()
  capitalized_words = []
  for word in words:
    if len(word) > 1:
      capitalized_word = word[0].upper() + word[1:-1] + word[-1].upper()
    else:
      capitalized_word = word.upper()
    capitalized_words.append(capitalized_word)
  return " ".join(capitalized_words)

text = s

capitalized_text = capitalize_first_last(text)

print(capitalized_text)
```
