```python
import pandas as pd # Read the CSV file into a pandas DataFrame
df = pd.read_csv("Employee_list.csv", index_col="Sno.")

# Create a dictionary to map professions to numeric codes
profession_map = {"Engineer": 0, "Doctor": 1, "Teacher": 2}

# Use the map() function to replace professions with codes
df["Profession"] = df["Profession"].map(profession_map)

# Add the profession to the employee's name
df["Name"] = df["Name"] + " (" + df["Profession"].astype(str) + ")"

print("**Answer 1:**\n", df)
```
