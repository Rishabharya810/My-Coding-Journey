```python
import pandas as pd

# **Qus. 1**

# Read the CSV file into a pandas DataFrame
df = pd.read_csv("Employee_list.csv", index_col="Sno.")

# Create a dictionary to map professions to numeric codes
profession_map = {"Engineer": 0, "Doctor": 1, "Teacher": 2}

# Use the map() function to replace professions with codes
df["Profession"] = df["Profession"].map(profession_map)

# Add the profession to the employee's name
df["Name"] = df["Name"] + " (" + df["Profession"].astype(str) + ")"

print("**Answer 1:**\n", df)
```

```python

# Calculate the increment for eligible employees
df["Increment"] = df["Salary"].apply(
    lambda salary: ((salary - 45000) * 100) / 45000
    if salary > 45000
    else 0
)

# Filter the DataFrame to select employees with salaries greater than 45000
selected_employees = df[df["Salary"] > 45000][
    ["Name", "Profession", "Salary", "Empid", "Increment"]
]

print("\n**Answer 2:**\n", selected_employees)
```
