```python
import pandas as pd

# Sample data (replace with your actual data)
data = {
    'Date': ['2023-10-26', '2023-10-27', '2023-10-28', '2023-10-26', '2023-10-27', '2023-10-28', '2023-10-26', '2023-10-27', '2023-10-28'],
    'Product': ['Product A', 'Product A', 'Product A', 'Product B', 'Product B', 'Product B', 'Product C', 'Product C', 'Product C'],
    'Sales': [10, 15, 12, 8, 11, 9, 13, 16, 14]
}
df = pd.DataFrame(data)

# 1. Pivot the DataFrame to wide format (if not already pivoted)
pivot_df = df.pivot(index='Date', columns='Product', values='Sales')

# 2. Melt the pivot_df back to long format
melted_df = pivot_df.melt(id_vars=['Date'], var_name='Product', value_name='Sales')

# Print the results
print("Original DataFrame:\n", df)
print("\nPivoted DataFrame:\n", pivot_df)
print("\nMelted DataFrame (back to long format):\n", melted_df)
```
