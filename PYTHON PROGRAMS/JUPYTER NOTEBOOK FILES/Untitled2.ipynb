```python
def sort_by_age(customers):
    """
    Sorts a list of customer ages in descending order (oldest to youngest)
    using Selection Sort without a temporary variable for swapping.

    Args:
        customers: A list of integers representing customer ages.

    Returns:
        The sorted list of customer ages.
    """

    n = len(customers)
    for i in range(n):
        max_index = i
        for j in range(i + 1, n):
            if customers[j] > customers[max_index]:
                max_index = j
        # Swap without temp variable using arithmetic operators
        customers[i], customers[max_index] = customers[i] + customers[max_index], customers[i]
        customers[max_index] = customers[i] - customers[max_index]
        customers[i] = customers[i] - customers[max_index] 
    return customers
customer = [23,54,2,12,45,65,67,87,54]
sorted_arr = sort_by_age(customer)
print(sorted_arr)
```
