# Linear Regression with Multiple Variables Practice

In this notebook, we'll practice implementing Linear Regression with multiple variables using a real dataset. We'll go through the following steps:
1. Data loading and inspection
2. Data preprocessing
3. Model training
4. Understanding the linear equation
5. Making predictions
6. Saving results

## 1. Import Required Libraries

First, let's import all the necessary libraries we'll need for this analysis.

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import os
import warnings
warnings.filterwarnings('ignore')
```

## 2. Load Dataset into Pandas DataFrame

Now we'll load our dataset from the CSV file and take a look at its contents.

```python
# Load the dataset
df = pd.read_csv('multiple_linear_regression_dataset.csv')

# Display the first few rows and basic information
print("First few rows of the dataset:")
print(df.head())
print("\nDataset Info:")
print(df.info())
```

## 3. Data Preprocessing: Handle Missing Values

Let's check for any missing values in our dataset and handle them if necessary.

```python
# Check for missing values
print("Missing values in dataset:")
print(df.isnull().sum())

# If there were missing values, we would handle them here
# But our dataset is clean, so we can proceed
```

## 4. Define Features and Target Variable

Let's separate our features (age and experience) from our target variable (income).

```python
# Separate features and target
X = df[['age', 'experience']]  # Features
y = df['income']  # Target

print("Features shape:", X.shape)
print("Target shape:", y.shape)

# Display first few rows of features
print("\nFirst few rows of features:")
print(X.head())
```

## 5. Train Linear Regression Model

Now we'll create and train our Linear Regression model.

```python
# Create and train the model
model = LinearRegression()
model.fit(X, y)

# Print the model's R-squared score
print("Model R-squared score:", r2_score(y, model.predict(X)))
```

## 6. Display Linear Equation

Let's see the coefficients and intercept to understand our linear equation.

```python
# Get coefficients and intercept
print("Coefficients:")
print(f"Age: {model.coef_[0]:.2f}")
print(f"Experience: {model.coef_[1]:.2f}")
print(f"Intercept: {model.intercept_:.2f}")

# Display the complete equation
print("\nLinear Equation:")
print(f"Income = {model.intercept_:.2f} + {model.coef_[0]:.2f}×Age + {model.coef_[1]:.2f}×Experience")
```

## 7. Make Predictions and Save Results

Let's make predictions using our model and save the results to a CSV file.

```python
# Make predictions
predictions = model.predict(X)

# Create a DataFrame with actual and predicted values
results_df = pd.DataFrame({
    'Age': X['age'],
    'Experience': X['experience'],
    'Actual_Income': y,
    'Predicted_Income': predictions,
    'Difference': y - predictions
})

# Display the first few rows of results
print("First few rows of results:")
print(results_df.head())

# Save to CSV file
output_file = 'multiple_linear_regression_predictions.csv'
results_df.to_csv(output_file, index=False)
print(f"\nResults saved to {output_file}")
```

## 8. Visualize Actual vs Predicted Income

Let's create a scatter plot to visualize how well our predictions match the actual income values.

```python
import matplotlib.pyplot as plt

# Set figure style to be clean and modern
plt.style.use('default')

# Create the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(results_df['Actual_Income'], results_df['Predicted_Income'], alpha=0.5, color='blue')
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r--', lw=2)  # Perfect prediction line

# Customize the plot
plt.title('Actual vs Predicted Income', fontsize=14)
plt.xlabel('Actual Income ($)', fontsize=12)
plt.ylabel('Predicted Income ($)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

# Add text box with R² score
r2 = r2_score(results_df['Actual_Income'], results_df['Predicted_Income'])
plt.text(0.05, 0.95, f'R² = {r2:.4f}', 
         transform=plt.gca().transAxes,
         fontsize=12,
         bbox=dict(facecolor='white', alpha=0.8))

plt.tight_layout()
plt.show()

# Create a residual plot
plt.figure(figsize=(10, 6))
residuals = results_df['Actual_Income'] - results_df['Predicted_Income']
plt.scatter(results_df['Predicted_Income'], residuals, alpha=0.5, color='blue')
plt.axhline(y=0, color='r', linestyle='--')

# Customize the residual plot
plt.title('Residual Plot', fontsize=14)
plt.xlabel('Predicted Income ($)', fontsize=12)
plt.ylabel('Residuals ($)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
